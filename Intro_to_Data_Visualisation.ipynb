### Types of Plots
1. **Line Chart**: Used when you want to show trends or make comparisons
2. **Bar Chart**: Used to display data grouped intro discrete non-numeric category
3. **Histogram**: Used to represent continous data and shape of data distribution
4. **Pie Chart**: Used to represent composition and show percentages
5. **Scatter Plot**: Used to find the relationship between 2 variables. Are also great for identifying outlier points and possible gaps in the data.
6. **Stacked Area Chart**: This chart is most often seen when combined with the concept of stacking, to show how both how a total has changed over time, but also how its componentsâ€™ contributions have changed.
